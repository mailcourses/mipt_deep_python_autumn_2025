```python
def create_logger(name):
    logger = logging.getLogger(name)
    logger.setLevel(logging.INFO)

    while logger.handlers:
        logger.removeHandler(logger.handlers[0])

    format1 = logging.Formatter("1111 %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")
    format2 = logging.Formatter("2222 %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")

    stream = logging.StreamHandler()
    stream.setLevel(logging.DEBUG)
    stream.setFormatter(format1)

    file_handler = logging.FileHandler(f"{name}.log", mode="w")
    file_handler.setLevel(logging.WARNING)
    file_handler.setFormatter(format2)

    logger.addHandler(stream)
    logger.addHandler(file_handler)

    return logger
```

```python
def create_logger(name):
    logger = logging.getLogger(name)
    logger.setLevel(logging.INFO)
    logger.propagate = False

    while logger.handlers:
        logger.removeHandler(logger.handlers[0])

    format1 = logging.Formatter("1111 %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")
    format2 = logging.Formatter("2222 %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")

    stream = logging.StreamHandler()
    stream.setLevel(logging.DEBUG)
    stream.setFormatter(format1)

    file_handler = logging.FileHandler(f"{name}.log", mode="w")
    file_handler.setLevel(logging.WARNING)
    file_handler.setFormatter(format2)

    logger.addHandler(stream)
    logger.addHandler(file_handler)

    return logger
```

```python
note2 = create_logger("note2")
note2.warning("warning99 in note2")
```

```python
class CustomFilter(logging.Filter):
    def filter(self, record):
        #print("RECORD", record.__dict__)
        return record.args[0] != "steve"
```

```python
note2.addFilter(CustomFilter())
```

```python
note2.warning("user %s connected to db %s", user, db)
```
